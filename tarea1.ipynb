```python
subprocess.call('curl "https://rest.kegg.jp/list/T00007" -o T00007.txt', shell = True) # lo hace en un cmd
```

```python
t00007 = []
with open('T00007.txt', 'r') as x:
    for fila in x:
        fila = fila.rstrip()
        t00007.append(fila.split('\t'))
```

```python
ncRNA = []
for i in t00007:
    if 'ncRNA' in i[1]:
        ncRNA.append(i)
```

```python
print(len(ncRNA), 'ncRNAs')
```

```python
n = 0
for i in t00007:
    if 'ncRNA' in i[1]:
        n += 1
print(n, 'ncRNAs')
```

```python
org = []
with open('kegg_organism_4.txt', 'r') as x:
    
    
    for fila in x:
        fila = fila.rstrip()
        org.append(fila.split('\t'))
```

```python
org
```

```python
org[:5]
```

```python
escherichia = []
for i in org:
    if 'Escherichia' in i[2]:
        escherichia.append(i[2])
```

```python
escherichia_species = []
for i in escherichia:
    escherichia_species.append(i.split(' ')[:2][0]+' '+i.split(' ')[:2][1])
```

```python
set(escherichia_species)
```

```python
list(set(escherichia_species))
```

```python
print(len(set(escherichia_species)), 'especies')
```

```python
ecoli = []
for i in escherichia:
    if 'Escherichia coli' in i:
        ecoli.append(i)
```

```python
len(set(ecoli))
```

```python
for i in list(set(escherichia_species)):
    if i != 'Escherichia coli':
        print(i)
```
