```python
def dna_a_arn2(a):
    l = []
    for i in a:
        if i == 'T':
            i = 'U'
            l.append(i)
        else:
            l.append(i)
    return ''.join(l)
```

```python
ADN = 'AAAGATCGATATCTCG'
```

```python
def rev_com(a):
    rev = ''.join(list(reversed(a)))
    l = []
    for i in rev:
        if i == 'A':
            i = 'T'
            l.append(i)
        elif i == 'C':
            i = 'G'
            l.append(i)
        elif i == 'G':
            i = 'C'
            l.append(i)
        elif i == 'T':
            i = 'A'
            l.append(i)
    return ''.join(l)
```

```python
rev_com(ADN)
```

```python
# Funcion de Yair de Jesus Gonzalez
def complemento_reverso (secuencia_de_adn):
    diccionario_del_complemento = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    complemento_reverso = ''

    for nucleotido in secuencia_de_adn:
        complemento_reverso = diccionario_del_complemento[nucleotido] + complemento_reverso
        #print(nucleotido,  diccionario_del_complemento[nucleotido] + complemento_reverso)

    return complemento_reverso
```

```python
print(ADN)
print(complemento_reverso (ADN))
```

```python
print(ADN)
print(complemento_reverso (ADN))
print(rev_com(ADN))
```

```python
# Funcion de Yair de Jesus Gonzalez
def com_rev1(a):
    dic_com = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    c_r = ''
    for n in ''.join(list(reversed(a))):
        c_r += dic_com[n]
    return c_r
```

```python
print(ADN)
print(complemento_reverso (ADN))
print(rev_com(ADN))
print(com_rev1(ADN))
```
